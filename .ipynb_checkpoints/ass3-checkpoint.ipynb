```python
import pandas as pd
import numpy as np


# Load dataset
df = pd.read_csv("/Users/meirrr/Desktop/data_analysis/hw3/nyc_temperature.csv")
df.head()
```

```python
# Data Wrangling
# Convert the 'date' column to datetime
df['date'] = pd.to_datetime(df['date'])

# Split by half of the year (e.g., first half and second half)
first_half = df[df['date'].dt.month <= 6]
second_half = df[df['date'].dt.month > 6]

# Concatenate
concatenated_data = pd.concat([first_half, second_half])
```

```python
# Covert from fahr to celsuis
df['tmax'] = (df['tmax'] - 32) * 5.0 / 9.0
df['tmin'] = (df['tmin'] - 32) * 5.0 / 9.0
df['tavg'] = (df['tavg'] - 32) * 5.0 / 9.0
```

```python
df.head()
```

```python
df.isnull()
```

```python
# Data Cleaning and Preparation
# Handle missing values
df.fillna(method='ffill', inplace=True)

# Remove duplicates
df.drop_duplicates(inplace=True)

# Set 'date' as index
df.set_index('date', inplace=True)

# Convert object data to numeric data
df['new_snow'] = pd.to_numeric(df['new_snow'], errors='coerce')
df['snow_depth'] = pd.to_numeric(df['snow_depth'], errors='coerce')
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
# Calculate a 7-day moving average
df['7_day_moving_avg'] = df['tavg'].rolling(window=7, min_periods=1).mean()

# Plot the rolling averages
plt.figure(figsize=(14, 6))
plt.plot(df.index, df['tavg'], label='Daily Average Temperature', alpha=0.5)
plt.plot(df.index, df['7_day_moving_avg'], label='7-Day Moving Average', color='orange')
plt.title('Daily Temperature with 7-Day and 30-Day Moving Averages')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.legend()
plt.show()
```

```python
# Monthly summary statistics
monthly_summary = df.resample('M').agg({
    'tavg': ['mean', 'std', 'min', 'max'],
    'tmax': ['mean', 'std', 'min', 'max'],
    'tmin': ['mean', 'std', 'min', 'max']
})
print("Monthly Summary Statistics:")
print(monthly_summary)
```
